```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import Normalizer


normal = np.random.rand(12,15)
transformer = Normalizer().fit(normal)
normal = transformer.transform(normal)
sns.heatmap(normal,cmap='autumn')
```

# Select the first breast cell line (AU565) in CCLE to draw heatmap for gene expression data

```python
path = "D:\Study\CIS\\CCLE\\"
df = pd.read_table(open(path + 'CCLE_gene_expression_train(breast).txt'), sep='\t')
data = list(df['AU565_BREAST'])




f, ax = plt.subplots(figsize=(8, 8))

temp = []
# print(type(data))
for i in range(0, 10):
    temp.append(data[5700*i: 5700*(i+1)])

temp = np.array(temp).T
print(temp.shape)
ax = sns.heatmap(temp, vmax=5, vmin=0, cmap='autumn')


plt.xlabel('Gene sets (containing 5700 genes each)',fontsize=20, color='k') #xlabel
plt.ylabel('Gene number',fontsize=20, color='k') #ylabel
cbar = ax.collections[0].colorbar
cbar.set_label('Transcriptome per million',fontdict={'size': 20})
plt.show()
```

# Filtered by cgc data set

```python
path = "D:\Study\CIS\\CCLE\\"
df = pd.read_table(open(path + 'CCLE_gene_expression_train(cgc).txt'), sep='\t')

data = list(df['AU565_BREAST'])
temp = []
# print(type(data))
for i in range(0, 5):
    temp.append(data[140*i: 140*(i+1)])

temp = np.array(temp).T
print(temp.shape)


f, ax = plt.subplots(figsize=(8, 8))
ax = sns.heatmap(temp, vmax=5, vmin=0, cmap='autumn')
plt.xlabel('Gene sets (containing 140 genes each)',fontsize=20, color='k') #xlabel
plt.ylabel('Gene number',fontsize=20, color='k') #ylabel
cbar = ax.collections[0].colorbar
cbar.set_label('Transcriptome per million',fontdict={'size': 20})
plt.show()
```
